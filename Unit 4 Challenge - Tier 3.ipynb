#### 1.2.  Loading the data

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices= "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
properties = properties.transpose()
properties.head()
```

```python
properties = properties.reset_index()
properties.columns = properties.iloc[0]
properties = properties.drop(properties.index[0])
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
properties.rename(columns={'Unnamed: 0': 'Borough', pd.NaT: 'ID'}, inplace=True)
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
tidy_properties = properties.melt(id_vars=['Borough', 'ID'], var_name='Month', value_name='Average_price')
tidy_properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Convert 'Average_price' to float using astype()
tidy_properties['Average_price'] = tidy_properties['Average_price'].astype(float)

# Convert 'Month' to datetime
tidy_properties['Month'] = pd.to_datetime(tidy_properties['Month'])
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Check for non-null observations in each column
tidy_properties.count()
```

```python
# Check the unique entries in the 'Borough' column
print(tidy_properties['Borough'].unique())

# Count the number of unique entries
print(len(tidy_properties['Borough'].unique()))
```

```python
# List of non-boroughs to be removed
non_boroughs = ['Inner London', 'Outer London', 'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
                'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
                'SOUTH WEST', 'England', 'City of London', 'Unnamed: 34', 'Unnamed: 37', 'Unnamed: 47']

# Filter the DataFrame to only include rows where the borough is not in the non_boroughs list
tidy_properties = tidy_properties[tidy_properties['Borough'].isin(non_boroughs) == False]


# Drop rows with NaN in 'ID' or 'Average_price'
tidy_properties = tidy_properties.dropna(subset=['ID', 'Average_price'])

# Confirm the number of unique boroughs
print(len(tidy_properties['Borough'].unique()))

# Confirm there are no NaN values in 'ID' and 'Average_price'
print(tidy_properties.isna().sum())
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
import matplotlib.pyplot as plt

# Choose a borough
borough_name = 'Camden'

# Subset the data for the chosen borough
borough_data = tidy_properties[tidy_properties['Borough'] == borough_name]

# Plotting
plt.plot(borough_data['Month'], borough_data['Average_price'])
plt.title('Average House Prices in ' + borough_name)
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
# Apply a lambda function to extract the year from each date in the 'Month' column
tidy_properties['Year'] = tidy_properties['Month'].apply(lambda x: x.year)

# Calculate the mean house prices for each year for each borough
borough_yearly_avg = tidy_properties.groupby(['Borough', 'Year']).mean(numeric_only=True)

# Reset index to make 'Borough' and 'Year' regular columns
borough_yearly_avg = borough_yearly_avg.reset_index()

# Define the start and end years
start_year = 1998
end_year = 2018

# Filter the DataFrame for the start and end years
start_prices = borough_yearly_avg[borough_yearly_avg['Year'] == start_year]
end_prices = borough_yearly_avg[borough_yearly_avg['Year'] == end_year]

# Merge the start and end prices DataFrames
merged_prices = pd.merge(start_prices, end_prices, on='Borough', suffixes=('_1998', '_2018'))

# Calculate the increase in price
merged_prices['Price_Increase'] = merged_prices['Average_price_2018'] - merged_prices['Average_price_1998']

# Calculate the percentage increase
merged_prices['Price_Percentage_Increase'] = (merged_prices['Price_Increase'] / merged_prices['Average_price_1998']) * 100

# Rank the boroughs by price increase
ranked_boroughs = merged_prices.sort_values(by='Price_Percentage_Increase', ascending=False)

# Display the top 5 boroughs with the greatest increase
print(ranked_boroughs[['Borough', 'Price_Percentage_Increase']].head())
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(dfg, borough_name):
    """
    Calculate the ratio of average house prices for a given borough between 1998 and 2018.
    
    """
    # Filter the data for the given borough
    borough_data = dfg[dfg['Borough'] == borough_name]
    
    # Get average prices for 1998 and 2018
    avg_price_1998 = borough_data[borough_data['Year'] == 1998]['Average_price'].mean()
    avg_price_2018 = borough_data[borough_data['Year'] == 2018]['Average_price'].mean()
    
    # Calculate the ratio of the average prices
    ratio = avg_price_1998 / avg_price_2018
    
    return ratio

# Test the function with 'Camden'
print(create_price_ratio(borough_yearly_avg, 'Camden'))
```

```python
# Dictionary to hold the boroughs and their respective price ratios
price_ratio_dict = {}

# Iterate over each unique borough
for borough in borough_yearly_avg['Borough'].unique():
    # Calculate the price ratio using the create_price_ratio function
    price_ratio = create_price_ratio(borough_yearly_avg, borough)
    # Store the result in the dictionary
    price_ratio_dict[borough] = price_ratio

# Convert the dictionary to a DataFrame for easier analysis and visualization
price_ratio_df = pd.DataFrame(list(price_ratio_dict.items()), columns=['Borough', 'Price_Ratio'])

# Display the boroughs with the highest price increase first
price_ratio_df.sort_values(by='Price_Ratio', ascending=True, inplace=True)
print(price_ratio_df.head())
```
